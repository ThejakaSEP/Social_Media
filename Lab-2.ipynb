```python
import pandas as pd
import numpy as np
import re

with open('YouTubeTrailerComments.txt',encoding='windows-1252') as f:
#     lines = f.readlines()
    lines = [line.rstrip('\n') for line in f.readlines()]
```

```python
len(lines)
```

```python
lines[0]
```

```python
pattern = r"MovieID::::\d+"
movie_id_list = []
movie_description = []
last_index = 0

for i,line in enumerate(lines):
    if re.match(pattern,line):
        colon_index = line.rindex(':')
        movie_id_list.append(line[colon_index+1:])
        movie_description.insert(-1,lines[last_index+1:i])
        last_index=i
        
print(last_index)
#insert the last description
movie_description.insert(-1,lines[last_index+1:])
```

```python
pattern_2 = r"\d+ \d+ \d+ \d+ \d+"
splitted_lines = []
for line in lines:
    if re.match(pattern_2,line):
        splitted_lines.append(line.split(' '))
```

```python
pattern_2 = r"\d+ \d+ \d+ \d+ \d+"
splitted_lines = []
for item in movie_description:
    flag=0
    for line in item:
        if re.match(pattern_2,line):
            splitted_lines.append(line.split(' '))
            flag = 1
    if flag!=1:
        splitted_lines.append([None]*5)
```
